```python
import pandas as pd
```

```python
# reference the data
data = "data/country_code.csv"
#this .csv was made by a teammate
```

```python
# Read our Data file with the pandas library
df = pd.read_csv(data)
#isolate the country codes
df = df.iloc[:3]
df
```

```python
#transpose the data
df = df.transpose()
df
```

```python
#make categories the columns
df.columns = df.iloc[0]
#drop first row
df.drop(index=df.index[0], 
        axis=0, 
        inplace=True)
```

```python
#make country the index
df = df.set_index(list(df)[0])
```

```python
#rename the columns
df = df.rename(columns ={ 'iso-a2':'iso_a2','iso-a3':'iso_a3'})
```

```python
#save clean .csv
df.to_csv('data/country_code_clean.csv')
```
